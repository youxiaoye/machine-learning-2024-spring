```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

x1 = np.arange(0, 20)
print(x1)

x2 = np.random.rand(20)
print(x2)

plt.scatter(x1, x2)
plt.show()

x1 = x1.reshape(4, 5)
print(x1.shape)

x1 = np.vstack(([1, 1, 1, 1, 1], x1))
print(x1)

x1 = np.insert(x1, 0, 2, axis=1)
print(x1)

def f(x):
    return 2*x**4 - 4*x + 9

result = minimize(f, x0=0)
print(result.fun)


```
