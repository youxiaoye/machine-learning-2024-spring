```python
#定义sigmoid函数
def sigmoid(z):
    return (1.0/(1.0+np.exp(-z)))
#定义前向传播函数
def feedforward_propagation(X,Theta1,Theta2):
    a1= np.insert(X,0,values= 1,axis =0)
    z2= np.dot(theta1,a1)
    a2= sigmoid(z2)
    a2= np.insert(a2,0,values= 1,axis =0)#再次添加偏置项
    z3= np.dot(theta2,a2)
    a3= sigmoid(z3)
    return a3
#定义预测函数
def predict(a3):
    return np.argmax(a3)+1
```

```python
#完成实际标签与预测标签的准确度对比的统计
def accuracy(X,y):
    sum_num = 0
    for i in range(y.size):
        if(y[i] == predict(feedforward_propagation(X[i],theta1,theta2))):
            sum_num += 1
    return sum_num/y.size
accuracy(X,y)
```
